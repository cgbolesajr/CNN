```python
import torch
import torchvision
from torchvision.transforms import transforms
```

```python
train_set = torchvision.datasets.FashionMNIST(
    root='./data'                                       # Extract
    ,train=True
    ,download=True
    ,transform=transforms.Compose([                     # Transform
        transforms.ToTensor()
    ])
)

train_loader = torch.utils.data.DataLoader(train_set,batch_size=10)
```

```python
grid = torchvision.utils.make_grid(images,nrow=10)
plt.figure(figsize=(15,15))
plt.imshow(grid.permute(1, 2, 0))
print('labels: ', labels)
```

 the input to matplotlib.pyplot.imshow() needs to be [image heigth x image width x # color channels] (i.e., the shape needs to be [518, 1292, 3]). The .permute(1, 2, 0) action is a Torch-specific function that permutes the axes of the original in exactly that order: [axis 1 x axis 2 x axis 0] = [image heigth x image width x # color channels].
